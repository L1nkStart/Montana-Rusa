# Paso 1: Método de Trazador Cúbico Sujeto

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Datos de ejemplo
xi = np.array([0, 1, 2, 3, 4])  
yi = np.array([1, 2, 0, 2, 3])  

# Condiciones de frontera (derivadas en los extremos)
d0 = 0  # Derivada en el primer punto
dn = 0  # Derivada en el último punto

# Creación del trazador cúbico con condiciones sujetas
cs = CubicSpline(xi, yi, bc_type=((1, d0), (1, dn)))

# Generación de valores para graficar
x_vals = np.linspace(xi[0], xi[-1], 100)
y_vals = cs(x_vals)

# Graficar los datos y la interpolación
plt.scatter(xi, yi, label="Puntos", color="red")
plt.plot(x_vals, y_vals, label="Trazador Cúbico Sujeto", color="blue", linewidth=2)
plt.xlabel("xi")
plt.ylabel("yi")
plt.title("Trazador Cúbico Sujeto")
plt.legend()
plt.grid()
plt.show()
```

```python
# Ajuste polinómico de grado 3 con polyfit
coef = np.polyfit(xi_2, yi_2, 3)  # Grado 3
p = np.poly1d(coef)

# Evaluamos el polinomio ajustado
x_vals_2 = np.linspace(min(xi_2), max(xi_2), 100)
y_vals_2 = p(x_vals_2)

# Graficamos
plt.scatter(xi_2, yi_2, color="red", label="Datos (xi, yi)")
plt.plot(x_vals_2, y_vals_2, color="blue", label=f"Ajuste polinómico grado 3")
plt.xlabel("Tensión")
plt.ylabel("Compresión")
plt.title("Ajuste por Mínimos Cuadrados")
plt.legend()
plt.grid()
plt.show()

# Mostrar ecuación ajustada
print(f"Polinomio ajustado: {p}")
```

## Paso 3: Polinomios Ortogonales

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de los polinomios de Legendre
def polinomio_legendre(n, x):
    if n == 0:
        return np.ones(len(x))
    elif n == 1:
        return x
    else:
        p_n_1 = polinomio_legendre(n - 1, x)
        p_n_2 = polinomio_legendre(n - 2, x)
        return ((2 * n - 1) * x * p_n_1 - (n - 1) * p_n_2) / n

# Valores de x
x = np.linspace(-1, 1, 400)

# Genera los polinomios de Legendre de grado 0 a 5
n_values = range(6)

valores_legendre = [polinomio_legendre(n, x) for n in n_values]

# Graficar los Polinomios de Legendre
plt.figure(figsize=(8, 6))

for i, n in enumerate(n_values):
    plt.plot(x, valores_legendre[i], label=f"P{n}(x)")

plt.xlabel("x")
plt.ylabel("Pn(x)")
plt.title("Polinomios de Legendre")
plt.legend()
plt.grid(True)
plt.ylim(-1.1, 1.1)
plt.show()
```

## Paso 4: Resolución de Ecuaciones

```python
import numpy as np
import sympy as sp
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir incógnitas
x1, x2, x3 = sp.symbols("x1 x2 x3")

# Crear el vector de incógnitas
v = sp.Matrix([x1, x2, x3])

# Leer el archivo CSV
df = pd.read_csv("Puntos_3.csv", header=None)

# Seleccionar las filas 2, 3 y 4 para la matriz A
matriz1 = df.iloc[1:4, :].values  

# Seleccionar la fila 5 como vector b
b = df.iloc[4, :].values.reshape(-1, 1)  

# Convertir a matrices SymPy
A = sp.Matrix(matriz1)
b = sp.Matrix(b)

# Mostrar las matrices
sp.pprint(A)
sp.pprint(b)

# Resolver el sistema de ecuaciones Ax = b
solucion = sp.linsolve((A, b), x1, x2, x3)

# Mostrar la solución
sp.pprint(solucion)

# Extraer valores de la solución
solucion_lista = list(solucion)[0]
valx1 = float(solucion_lista[0])
valx2 = float(solucion_lista[1])
valx3 = float(solucion_lista[2])

# Crear gráfico 3D del punto crítico
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter([valx1], [valx2], [valx3], color='red', s=100)

# Etiquetas
ax.set_title("Punto Crítico")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Mostrar gráfico
plt.show()
```
